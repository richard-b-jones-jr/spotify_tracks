# Spotify Tracks - Exploratory Data Analysis (EDA)

Data Source: https://www.kaggle.com/datasets/maharshipandya/-spotify-tracks-dataset

This is a dataset of Spotify tracks over a range of 125 different genres. Each track has some audio features associated with it.

What can be learned from this dataset?
- what combinations of factors do the most (and least) popular songs have in common?
- What combinations of factors are extremely uncommon, and may be possible inroads to a unique new song idea?
- How popular are my favorite songs, albums and artists? (Am I hip?)

#### Column Descriptions
- track_id: 
    - The Spotify ID for the track
    
    
- artists: 
    - The artists' names who performed the track. If there is more than one artist, they are separated by a ;
    
    
- album_name:
    - The album name in which the track appears
    
    
- track_name:
    - Name of the track
    
    
- popularity: 
    - The popularity of a track is a value between 0 and 100, with 100 being the most popular. The popularity is calculated by algorithm and is based, in the most part, on the total number of plays the track has had and how recent those plays are. Generally speaking, songs that are being played a lot now will have a higher popularity than songs that were played a lot in the past. Duplicate tracks (e.g. the same track from a single and an album) are rated independently. Artist and album popularity is derived mathematically from track popularity.
    
    
- duration_ms: 
     - The track length in milliseconds
     
     
- explicit:
    - Whether or not the track has explicit lyrics (true = yes it does; false = no it does not OR unknown)
    
    
- danceability: 
    - Danceability describes how suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity. A value of 0.0 is least danceable and 1.0 is most danceable
    
    
- energy: 
    - Energy is a measure from 0.0 to 1.0 and represents a perceptual measure of intensity and activity. Typically, energetic tracks feel fast, loud, and noisy. For example, death metal has high energy, while a Bach prelude scores low on the scale


- key: 
    - The key the track is in. Integers map to pitches using standard Pitch Class notation. E.g. 0 = C, 1 = C♯/D♭, 2 = D, and so on. If no key was detected, the value is -1


- loudness: 
    - The overall loudness of a track in decibels (dB)


- mode: 
    - indicates the modality (major or minor) of a track, the type of scale from which its melodic content is derived. Major is represented by 1 and minor is 0


- speechiness: 
    - detects the presence of spoken words in a track. The more exclusively speech-like the recording (e.g. talk show, audio book, poetry), the closer to 1.0 the attribute value. Values above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks


- acousticness: 
    - A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic


- instrumentalness: 
    - Predicts whether a track contains no vocals. "Ooh" and "aah" sounds are treated as instrumental in this context. Rap or spoken word tracks are clearly "vocal". The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content


- liveness: 
    - Detects the presence of an audience in the recording. Higher liveness values represent an increased probability that the track was performed live. A value above 0.8 provides strong likelihood that the track is live


-  valence: 
    - A measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry)


- tempo: 
    - The overall estimated tempo of a track in beats per minute (BPM). In musical terminology, tempo is the speed or pace of a given piece and derives directly from the average beat duration


- time_signature: 
    - An estimated time signature. The time signature (meter) is a notational convention to specify how many beats are in each bar (or measure). The time signature ranges from 3 to 7 indicating time signatures of 3/4, to 7/4.


-  track_genre: 
    - The genre in which the track belongs

## Feature Engineering

There are several features that need tweaks. 
- Key and Mode are currently continuous but need to be treated as categorical.
- duration will make more sense in minutes than milliseconds

New features can be created to help understand the dataset.
- Combining key and mode could be useful together moreso than separately.
- Creating bins by quartile for Popularity could help to visualize distributions categorical factors.

```python
# Convert duration into minutes from ms
df['duration'] = (df['duration_ms'] / (1000 * 60)).astype(float)

# Drop ms column
df.drop(columns='duration_ms',inplace=True)
```

```python
categorical_columns = df.select_dtypes(include=['object','bool']).columns.tolist()

# Add time_signature to categorical list
categorical_columns.extend(['time_signature'])
```

```python
# Convert columns to categorical type
for c in categorical_columns:
    df[c] = df[c].astype('category')
    
# Revise explicit to binary type for ease of use
df['explicit'] = np.where(df['explicit']==True,1,0)
    
# Map in actual key values
key_dict = {0:'C',
            1:'C#',
            2:'D',
            3:'D#',
            4:'E',
            5:'F',
            6:'F#',
            7:'G',
            8:"G#",
            9:'A',
            10:'A#',
            11:'B'}

# Map in actual mode values
mode_dict = {0:'Minor',1:'Major'}

# Create column to replace numeric key with the actual key
df['mapped_key'] = df['key'].map(key_dict)

# Create column to replace numeric mode with the actual mode
df['mapped_mode'] = df['mode'].map(mode_dict)

# Create a column combining key and mode
df['key_mode'] = (df['mapped_key'].astype(str) + ' ' + df['mapped_mode'].astype(str)).astype('category')
```

```python
# Create quartiles by popularity
quartile_labels = ['Low','Medium','High','Extreme']
df['popularity_bin'] = pd.qcut(df['popularity'], q=4, labels=quartile_labels)
```

#### Further Check Data Accuracy

As a musician, I know the key of many of my favorite songs. Is the key and mode in the dataset accurate?

```python
my_faves = ['Queens of the Stone Age','The Beatles','AC/DC']
df[df['artists'].isin(my_faves)][['artists','track_name','key_mode']].sample(10,random_state=6)
```

Less than half of these songs have the correct key, and the others are way off. I'm going to remove the drop the key and mode columns, since valence should convey a similar meaning and be more reliable.

```python
df.drop(columns=['key','mode','mapped_key','mapped_mode','key_mode'], inplace=True)
```

#### Are there multiple instances of the same song?

Songs can be on soundtracks and albums. If popularity is to be predicted, each song should only appear once per artist.

```python
artist_track_grp = df.groupby(['artists','track_name'])
df['track_count'] = artist_track_grp['track_name'].transform('count')
```

```python
df[(df['artists']=='AC/DC') & (df['track_count']>1)].sort_values(by='track_name').head(7)
```

Sure enough, one song can show up multiple times, sometimes completely duplicated, other times with only a different genre, or from a soundtrack iteration.

```python
single_artist_track_count = df[df['track_count']==1]['track_name'].count()
multi_artist_track_count = df[df['track_count']>1]['track_name'].count()


print(f'{single_artist_track_count:,.0f} unique tracks. {multi_artist_track_count:,.0f} duplicate songs found')

# Remove track_count
df.drop(columns={'track_count'},inplace=True)
```

#### Genre Consolidation

Many artists' songs show up with multiple genres. I'll select the genre that is most often associated with each artist before removing the extra rows.

```python
# Count genres per artist
top_genre = df.groupby(['artists','track_genre']).agg(records=('track_genre','count'))

# Select top most frequent genre per artist
top_genre = top_genre[top_genre > 0].sort_values(by='records',ascending=False).groupby('artists').head(1).reset_index()

# Remove aggregation
top_genre.drop(columns={'records'},inplace=True)

# Rename columns
top_genre.columns = ['artists','artist_genre']

# Merge artist genre 
df3 = df.merge(how='left', right=top_genre, on='artists')

# View Sample
df3[['artists','track_name','popularity','valence','duration','artist_genre']].sample(5,random_state=92)
```

Now that genres are consolidated, duplicate tracks will be dropped.

```python
df1 = df3.drop_duplicates(['artists','track_name'])
print(f'{len(df3)-len(df1):,.0f} of {len(df):,.0f} rows removed.')
```

## Feature Distributions

### Categorical variables

```python
# Define re-usable function to plot categorical distributions
def categorical_count(df, column, color_by_popularity=False):
    
    # Get descending order of index 
    order_descending = df[column].value_counts().index
    
    # Create figure
    fig, ax = plt.subplots(figsize=(10, 4))
    
    if color_by_popularity:
        sns.countplot(data=df,y=column,order=order_descending,hue='popularity_bin')
    else:
        sns.countplot(data=df,y=column,order=order_descending)
        
    fig.suptitle(f'Distribution of {column}')
```

```python
categorical_count(df1,'time_signature')
```

```python
categorical_count(df1,'explicit')
```

Most songs are in 4/4 and don't include explicit language.

#### Distribution of categoricals with respect to popularity

```python
def stacked_bar_categorical_popularity(df, column, independent_variable, display_table=False):
    
    # Create a stacked percentage barplot
    plt.figure(figsize=(10, 4))
    sns.histplot(
        data=df,
        y=column,
        hue=independent_variable,
        multiple="fill", 
        stat="proportion",
        discrete=True,
        shrink=.8

    )
    plt.title(f"{independent_variable} distribution by {column}")
    plt.show()

    if display_table:
        counts = df.groupby([column, independent_variable]).size().reset_index(name='count')
        total_counts = df.groupby(column).size().reset_index(name='total_count')
        merged_df = pd.merge(counts, total_counts, on=column)
        merged_df['percentage'] = (merged_df['count'] / merged_df['total_count']) * 100
        display(merged_df)
```

```python
stacked_bar_categorical_popularity(df=df1, column='explicit',independent_variable='popularity_bin')
```

As percent of total, explicit songs have a small edge on non-explicit songs for popularity.

```python
stacked_bar_categorical_popularity(df=df1, column='time_signature',independent_variable='popularity_bin')
```

Songs in 4/4 are generally most popular.

## Continuous Variable Distributions

```python
def hist_and_box_plot(df, variable):
    
    fig, axes = plt.subplots(3, 1, figsize=(9, 5),sharex=True)

    # Box plot
    sns.boxplot(x=variable, data=df, ax=axes[0],meanline=True,showmeans=True)
    axes[0].xaxis.label.set_visible(False)

    # Histogram
    sns.histplot(data=df, x=variable, kde=True, ax=axes[1])
    # show average 
    avg_var_val = np.mean(df[variable])
    axes[1].axvline(avg_var_val, color='green', linestyle='--', label=f'Overall Average: {avg_var_val:.2f}')
    
    # hide labels
    axes[1].xaxis.label.set_visible(False)
    axes[1].yaxis.label.set_visible(False)
    
    # Distribution of popularity within the variable
    sns.kdeplot(data=df, x=variable,hue='popularity_bin')
    axes[2].xaxis.label.set_visible(False)
    axes[2].yaxis.label.set_visible(False)
    
    plt.tight_layout()
    plt.suptitle(f'{variable}', y=1.02)
    plt.show()
```

```python
continuous_columns = df.select_dtypes(include=['int64','float64']).columns.tolist()

for c in continuous_columns:
    hist_and_box_plot(df1,c)
    print('-' * 124)
```

```python
hist_and_box_plot(df1[(df1['duration']>1) & (df1['duration']<=7)],'duration')
```
